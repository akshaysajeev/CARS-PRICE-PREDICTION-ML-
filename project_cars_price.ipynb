```python


# getdummy using pandas
# to change the cotogerical string data into dummy  sets
dummy=pd.get_dummies(df[['Location','Fuel_Type','Transmission','Owner_Type']],drop_first=True)
dummy
```

```python
# combine 2 dataframes to get kilometer,mileage,etc from df and encoded df from dummy
# use concat
df1=pd.concat([df,dummy],axis=1)
df1
```

```python
# drop the unwanted cols from this
df_final=df1.drop((['Unnamed: 0', 'Name', 'Location','Fuel_Type', 'Transmission', 'Owner_Type','New_Price']),axis=1)
df_final
```

```python
# to remove the string values inside columns like kmpl in mileage column
# function:  str.replace
df_final['Mileage']=df_final['Mileage'].str.replace('km/kg','')
df_final['Mileage']=df_final['Mileage'].str.replace('kmpl','')
df_final['Power']=df_final['Power'].str.replace('bhp','')
df_final['Engine']=df_final['Engine'].str.replace('CC','')
df_final
```

```python
# to remove 'null' in mileage replaced with 0
df_final['Mileage']=df_final['Mileage'].str.replace('null','0')
df_final['Power']=df_final['Power'].str.replace('null','0') 
df_final['Engine']=df_final['Engine'].str.replace('null','0')
```

```python
df_final.isna().sum()
```

```python
df_final.dtypes
```

```python
# in a machine learning model all columns should be integer or float(numerical values)
#  convert object type to float
df_final['Mileage']=df_final['Mileage'].astype(float)
df_final['Engine']=df_final['Engine'].astype(float)
df_final['Power']=df_final['Power'].astype(float)
df_final.dtypes
```

```python
# convert the 0 assigned in mileage ,engine ,power to NAN(not a number) 
df_final.loc[df_final.Engine==0,'Engine']=np.NaN
df_final.loc[df_final.Power==0,'Power']=np.NaN
df_final.loc[df_final.Mileage==0,'Mileage']=np.NaN
```

```python
df_final.isna().sum()
```

```python
#fill the missing values
# fill engine power mileage with mean 
# fill seat with mode
df_final['Engine']=df_final['Engine'].fillna(df_final['Engine'].mean())
df_final['Mileage']=df_final['Mileage'].fillna(df_final['Mileage'].mean())
df_final['Power']=df_final['Power'].fillna(df_final['Power'].mean())
df_final['Seats']=df_final['Seats'].fillna(df_final['Seats'].mode()[0])
df_final.isna().sum()
```

```python
# the pure final data for traing
df_final
```

```python
# assigning traing data
# x_train
x=df_final.drop(['Price','Fuel_Type_Electric'],axis=1)
x
# dropping fuel type electric because of in testing data there is no category fuel type electric
```
